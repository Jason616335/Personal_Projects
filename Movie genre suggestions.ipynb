## Visualization 4: The spread of different rating about high score movies.
This is the most attractive part. I want to know if high score movies have a relationship with big gross and big gross. Now I have a data set called subset5 that has high score and high return. But I want to know if film classfication influences score and return. So I use groupby("rating") first and then use describe() to show the classification of rating and get the numer of each rating. 

```python
subset5.groupby('rating').describe()
```

There are only five ratings. It reduces from totally 8 to 5. But the ratings are string. They can't be calculated and display in a scatter plot. So I use for loop and if conditional statement to make them look like a number.

I assign the new data to a data set called subset6.

```python
#This is 2015-2018 avocado average price in SF.
subset6=subset5
for i in range(23):
    rate=subset5.iloc[i,7]
    
    if (rate=="G"):
        subset6.iloc[i,7]=0

    if (rate=="NC-17"):
        subset6.iloc[i,7]=1
        
    if (rate=="NOT RATED"):
        subset6.iloc[i,7]=2
        
    if (rate=="Not specified"):
        subset6.iloc[i,7]=3
        
    if (rate=="PG"):
        subset6.iloc[i,7]=4
        
    if (rate=="PG-13"):
        subset6.iloc[i,7]=5
        
    if (rate=="R"):
        subset6.iloc[i,7]=6
        
    if (rate=="UNRATED"):
        subset6.iloc[i,7]=7
```

```python
subset6.head()
```

There is a problem of rating. Even though gross column, budget column and rating column look like numeric values, ratings are floating, which means can't be calculated by color in scatter plot function. So I make them into int values by using astype(int). Then I assign them into scatter plot to show plot.

```python
subset6['gross']=subset6['gross'].astype(int)
subset6['budget']=subset6['budget'].astype(int)
subset6['rating']=subset6['rating'].astype(int)

fig = plt.figure(figsize=(10,10))

colormap = np.array([ 'red', 'blue', 'green', 'yellow','orange','pink','grey','black'])
 
plt.scatter(x=subset6.gross, y=subset6.budget, c = colormap[subset6.rating], s = 50)

plt.ylabel("The budget of a movie ",fontsize=20)
plt.xlabel("The gross of a movie",fontsize=20)
plt.title('The spread of different rating movies',fontsize=20)
plt.show()
```


Rating -- colors：

G -- red

NC-17 -- blue

NOT RATED -- green

Not specified -- yellow

PG -- orange

PG-13 -- pink

R -- grey

UNRATED -- black
